```python
platform_totals=df[['Netflix','Hulu']].sum().sort_values(ascending=False)

#separating the values
platforms=platform_totals.index 
counts=platform_totals.values


plt.barh(platforms,counts,height=0.5)

plt.title="Total Content by Streaming Platform"
plt.xlabel="Number of Titles"
plt.ylabels="Platform"
#to invert the bar graph
plt.gca().invert_yaxis()


plt.show()
```

```python
platform_totals=df[['Netflix','Hulu']].sum().sort_values(ascending=False)

#separating the values
platforms=platform_totals.index 
counts=platform_totals.values

plt.subplot(2,1,1)

plt.barh(platforms,counts,height=0.5)

plt.title="Total Content by Streaming Platform"
plt.xlabel="Number of Titles"
plt.ylabels="Platform"
#to invert the bar graph
plt.gca().invert_yaxis()

mylabels = ["Netflix ", "Prime Video", "Hulu  ", "Disney+ "]
myexplode=[0.2,0,0,0]
plt.subplot(2,1,2)

plt.pie(platform_totals,labels=mylabels,explode=myexplode,autopct='%1.1f%%')
plt.legend(title="Platforms")

plt.show()
```
